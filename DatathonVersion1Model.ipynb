```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
df1 = pandas.read_csv('vehicle.csv')
```

```python
from sklearn import cluster 
```

```python
##Method 1 K.Mean to group
```

```python
model = cluster.KMeans(n_clusters = 5, random_state=14)
```

```python
fit = model.fit(df1)
```

```python
df1["cluster"] = fit.predict(df1)
```

```python
#df1["target"] = df1.
```

```python
df1['c'] = "count"
```

```python
clustering_result = df1[["cluster", "c"]].groupby("cluster").agg("count")
print(clustering_result)
```
